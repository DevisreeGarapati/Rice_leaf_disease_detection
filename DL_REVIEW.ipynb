```python
dict={"bacteria":list(data_dir.glob("bacterial_leaf_blight/*")),
      "brown":list(data_dir.glob("Brown_spot/*")),
      "smut":list(data_dir.glob("Leaf Streak/*")),
     "Healthy":list(data_dir.glob("Healthy/*")),
     "blast":list(data_dir.glob("leaf_blast/*")),
     "hispa":list(data_dir.glob("Hispa/*")),
     "scald":list(data_dir.glob("leaf_scald/*")),
      "Shath Blight":list(data_dir.glob("Shath Blight/*")),
      "narrow":list(data_dir.glob("narrow_brown_spot/*")),
      "Tungro":list(data_dir.glob("Tungro/*"))
     }
```

```python
labels_dict = {
    'bacteria': 0,
    'brown': 1,
    'smut': 2,
    'Healthy': 3,
    'blast': 4,
    'hispa': 5,
    'scald': 6,
    'Shath Blight': 7,
    'narrow': 8,
    'Tungro': 9
}
```

```python
str(dict["smut"][0])
```

```python
img=cv2.imread(str((dict["smut"][0])))
```

```python
cv2.resize(img,(180,180)).shape
```

```python
X, y = [], []
for name, images in dict.items():
    for image in images:
        img = cv2.imread(str(image))
        resized_img = cv2.resize(img,(180,180))
        X.append(resized_img)
        y.append(labels_dict[name])
```

```python
y[:5]
```

```python
X = np.array(X)
y = np.array(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
len(X_test)
```

```python
X_train_scaled = X_train / 255
X_test_scaled = X_test / 255
```

```python
#VGG19 with 16 convolutiona layers and 3 fully connected layers
model = Sequential([
    layers.Conv2D(16,(3,3),activation = 'relu',input_shape = (180,180,3)),
    layers.MaxPooling2D(2,2),
    layers.Conv2D(32,(3,3),activation = 'relu'),
    layers.MaxPooling2D(2,2),
    layers.Conv2D(64,(3,3),activation = 'relu'),
    layers.MaxPooling2D(2,2),
    layers.Flatten(),
    layers.Dense(128,activation = 'relu'),
    layers.Dropout(0.2),
    layers.Dense(256,activation = 'relu'),
    layers.Dropout(0.3),
    
    layers.Dense(4,activation = 'softmax')

],    name = 'Conv2D_Model')
```

```python
data_augmentation = keras.Sequential(
  [

    layers.experimental.preprocessing.RandomZoom(0.2),
    layers.experimental.preprocessing.RandomRotation(0.1),
    layers.experimental.preprocessing.RandomFlip("horizontal")
  ]
)
num_classes = 10

model = Sequential([
  data_augmentation,
  layers.Conv2D(16, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(32, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(64, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Dropout(0.1),
  layers.Flatten(),
  layers.Dense(128, activation='relu'),
  layers.Dense(num_classes)
])

model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
              
model.fit(X_train_scaled, y_train, epochs=10)
```

```python
model.evaluate(X_test_scaled,y_test)
```
